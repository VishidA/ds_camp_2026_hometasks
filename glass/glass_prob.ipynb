```python
def solve(floors, glasses, memo=None):
    if memo is None:
        memo = {}

    if(floors, glasses) in memo:
        return memo[(floors, glasses)]

    if glasses == 0 or floors == 0:
        return 0
    
    if glasses == 1:
        return floors
    
    answ = float('inf')

    for i in range(1, floors + 1):
        temp = max(solve(i-1, glasses-1, memo), solve(floors-i, glasses, memo)) + 1
        answ = min(answ, temp)

    memo[(floors, glasses)] = answ

    return answ
```

```python
test_cases = [
    (10,2),
    (100,2),
    (100,3),
    (1000,3),
    (1000,5)
]

for floors, glasses in test_cases:
    print(f"answer for {floors} floors and {glasses} glasses:", solve(floors, glasses))
```
